```python
def temperature_conversion(temp, unit):
    """Convert the Temperature from Fahrenheit to Celsius"""
    if unit=="c":
        return temp * 9/5 + 32
    elif unit=="f":
        return (temp - 32) * 5/9
    else:
        return None
    
print(temperature_conversion(15, "c"))
print(temperature_conversion(19, "f"))
```
