```python
class BankAccount:
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        
    def deposit (self, amount):
        self.balance+=amount
        print(f"The amount deposited = {amount} The new balance = {self.balance}")
        
    def withdraw (self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance-=amount
            
        print(f"The amount withdrawn = {amount} The new balance = {self.balance}")
        
    def balance(self):
        return self.balance
        
account = BankAccount("Momohjimoh", 50000)
print(account.balance)
```
